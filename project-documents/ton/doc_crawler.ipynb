```python
import requests

import requests

url = "https://docs.ton.org/participate/"

try:
    # 发送GET请求
    response = requests.get(url)

    # 检查响应状态码
    if response.status_code == 200:
        # 打印网页内容
        print(response.text)
    else:
        print("请求失败，状态码：", response.status_code)

except requests.exceptions.RequestException as e:
    print("请求发生异常:", e)
```
